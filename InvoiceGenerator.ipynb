```python
import random
import os
from faker import Faker
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from datetime import datetime
```

```python
# Inicializa Faker
fake = Faker()
```

```python
# Crear carpeta "invoices" si no existe
if not os.path.exists('invoices'):
    os.makedirs('invoices')
```

```python
supplier_expense_mapping = {
    "HarvestEdge Solutions": "Maintenance",
    "SanJo Plastics": "Packaging Materials",
    "Prime Distributors": "Transportation and Logistics",
    "Aromatic Essentials Ltd.": "Cleaning Services",
    "Mccall Group": "Repairs and Maintenance",
    "FreshClove Ventures": "Utilities",
    "PureGarlic Goods": "Raw Garlic Purchases",
    "Vital Group": "Quality Control",
    "BulbPrime Processing": "Pest Control",
    "GoldenRoots LLC": "Safety Gear"
}

# Lista de suppliers (extraída del diccionario para consistencia)
suppliers = list(supplier_expense_mapping.keys())

# Función para generar una factura
def generate_invoice(invoice_number, year):
    # Datos aleatorios
    client_name = "7 De Oro"
    client_address = "730 Manuel Ortiz Guerrero, San Lorenzo, Paraguay"
    #client_name = fake.name()
    supplier = random.choice(suppliers)  # Escoger un supplier de la lista
    expense_type = supplier_expense_mapping[supplier]  # Obtener el expense_type consistente del diccionario
    supplier_address = fake.address().replace("\n", ", ")
    amount = round(random.uniform(100, 1000), 2)
    
    # Generar una fecha aleatoria
    month = random.randint(1, 12)
    day = random.randint(1, 28)
    date_of_invoice = f"{day:02d}-{month:02d}-{year}"
    
    # Crear carpeta por mes dentro de "invoices"
    month_folder = f"invoices/{month:02d}-{year}"
    if not os.path.exists(month_folder):
        os.makedirs(month_folder)
    
    # Crear el archivo PDF en la carpeta correspondiente
    pdf_filename = f'{month_folder}/Invoice_{invoice_number}.pdf'
    c = canvas.Canvas(pdf_filename, pagesize=letter)

    # Título de la factura
    c.setFont("Helvetica-Bold", 20)
    c.drawString(200, 720, f"Invoice #{invoice_number}")  # Bajado de 730

    # Información de la empresa
    c.setFont("Helvetica", 16)
    c.drawString(30, 680, f"Supplier: {supplier}")       # Bajado de 700
    c.drawString(30, 650, f"Address: {supplier_address}") # Bajado de 685

    # Información del cliente
    c.drawString(30, 600, f"Client: {client_name}")       # Bajado de 630
    c.drawString(30, 570, f"Client Address: {client_address}")  # Bajado de 615

    # Información de la factura
    c.drawString(30, 520, f"Date: {date_of_invoice}")     # Bajado de 585
    c.drawString(30, 490, f"Expense Type: {expense_type}")  # Bajado de 570
    c.drawString(30, 460, f"Amount: ${amount}")           # Bajado de 555

    # Pie de página con información adicional
    c.setFont("Helvetica", 8)
    c.drawString(30, 50, "This invoice has been randomly generated for educational purposes only.")  # No modificado
        
    # Guardar el PDF
    c.save()
    return pdf_filename
```

```python
# Generar facturas para el año 2024
year = 2024
for invoice_number in range(1, 201):
    generate_invoice(invoice_number, year)
    print(f"Invoice {invoice_number} created.")
```
